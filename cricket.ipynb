```python
import numpy as np
import pandas as pd
data =pd.read_csv('IPL Ball-by-Ball 2008-2020.csv')
dataByMatch=pd.read_csv("IPL Matches 2008-2020.csv")
data.head(3)
```

```python
data['ballNo']=data.apply(lambda x: [1], axis=1, result_type='expand' )
data['costRate']=data.apply(lambda x:[1],axis=1, result_type='expand');
data1=data.groupby(['id','batsman']).sum()
# data.head(5)
```

```python
data1['runRate']=data1.apply(lambda x:(x['batsman_runs']/x['ballNo'])*100,axis=1, result_type='expand')
data1.head(3)
```
