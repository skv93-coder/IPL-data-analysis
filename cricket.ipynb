```python
import numpy as np
import pandas as pd
data =pd.read_csv('IPL Ball-by-Ball 2008-2020.csv')
dataByMatch=pd.read_csv("IPL Matches 2008-2020.csv")
```

```python
data['ballNo']=data.apply(lambda x: [1], axis=1, result_type='expand' )
data1=data.groupby(['id','batsman']).sum()
data.head(5)
```

```python
data1['runRate']=data1.apply(lambda x:(x['batsman_runs']/x['ballNo'])*100,axis=1, result_type='expand')
data1.head(19)
```

```python
data2=data.groupby(['id','bowler']).sum()
data2.head(19)
```

```python
data2['costRate']=data2.apply(lambda x:((x['is_wicket']/x['ballNo'])*100 if x['total_runs']==0 else x['ballNo']/x['total_runs']+x['is_wicket']/x['ballNo'])*100,axis=1, result_type='expand')
data2.head(19)
```
